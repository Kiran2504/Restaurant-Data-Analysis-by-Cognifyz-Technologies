```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Data
df = pd.read_csv("Dataset .csv")
```

__Task_1__

```python
# --- TASK 1: RESTAURANT RATINGS ---
print("--- Task 1: Restaurant Ratings ---")
# Histogram
plt.figure(figsize=(8, 5))
df['Aggregate rating'].hist(bins=20, color='teal', edgecolor='black')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Most common range (Using bins)
bins = [0, 1, 2, 3, 4, 5]
rating__counts = pd.cut(df['Aggregate rating'], bins=bins).value_counts().sort_values(ascending=False)
print("Most common rating range:", rating__counts.idxmax())
print(f"Average Votes: {df['Votes'].mean():.2f}")
```

__Task-2__

```python
# --- TASK 2: CUISINE COMBINATION ---
print("\n--- Task 2: Cuisine Combination ---")
cuisinecombos = df['Cuisines'].value_counts().head(5)
print("Top 5 Cuisine Combinations:")
print(cuisinecombos)

top_comboslist = cuisinecombos.index.tolist()
avg_rating_combos = df[df['Cuisines'].isin(top_comboslist)].groupby('Cuisines')['Aggregate rating'].mean()
print("\nAverage Ratings for these combinations:")
print(avg_rating_combos)
```

__Task-3__

```python
# --- TASK 3: GEOGRAPHIC ANALYSIS ---
print("\n--- Task 3: Geographic Analysis ---")
plt.figure(figsize=(10, 6))
plt.scatter(df['Longitude'], df['Latitude'], alpha=0.5, s=10, c='orange')
plt.title('Restaurant Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
print("Map plotted successfully.")
```

__Task-4__

```python
# --- TASK 4: RESTAURANT CHAINS ---
print("\n--- Task 4: Restaurant Chains ---")
restaurant__counts = df['Restaurant Name'].value_counts()
chains = restaurant__counts[restaurant__counts > 1]
print(f"Are chains present? {'Yes' if not chains.empty else 'No'}")

# Stats for Top 5 Chains
top_5_chains = chains.head(5).index.tolist()
chain_stats = df[df['Restaurant Name'].isin(top_5_chains)].groupby('Restaurant Name').agg({
    'Aggregate rating': 'mean',
    'Votes': 'sum'
})
print("\nTop 5 Chains Stats:")
print(chain_stats)
```
