```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Load the dataset
# Make sure "Dataset .csv" is in the same folder
df = pd.read_csv("Dataset .csv")
```

```python
print("Data Loaded Successfully!\n")
```

__Task-1__

```python
# --- TASK 1: TOP CUISINES ---
print("--- Task 1: Top Cuisines ---")
# Handling missing values and splitting cuisines in dataset
cuisines_data = df['Cuisines'].dropna()
all_cuisines = []
for row in cuisines_data:
    cuisines = [c.strip() for c in row.split(',')]
    all_cuisines.extend(cuisines)

cuisine_counts = pd.Series(all_cuisines).value_counts()
top_cuisines_list = cuisine_counts.head(3)
total_restaurants = len(df)

print("Top_3_Cuisines List:")
print(top_cuisines_list)

print("\nPercentages:")
for cuisine, count in top_cuisines_list.items():
    percentage = (count / total_restaurants) * 100
    print(f"{cuisine}: {percentage:.2f}%")
```

__Task-2__

```python
# --- TASK 2: CITY ANALYSIS ---
print("\n--- Task 2: City Analysis ---")
city_counts = df['City'].value_counts()
top_city = city_counts.idxmax()
print(f"City with most restaurants: {top_city} ({city_counts.max()} restaurants)")

city_avg_ratings = df.groupby('City')['Aggregate rating'].mean()
highest_rated_city = city_avg_ratings.idxmax()
print(f"City with highest average rating: {highest_rated_city} ({city_avg_ratings.max():.2f})")
```

__Task-3__

```python
# --- TASK 3: PRICE RANGE DISTRIBUTION ---
print("\n--- Task 3: Price Range Distribution ---")
price_range_counts = df['Price range'].value_counts().sort_index()
price_range_percentages = (price_range_counts / total_restaurants) * 100
print("Price Range Percentages:")
print(price_range_percentages)
```

```python
# Plotting
plt.figure(figsize=(8, 5))
price_range_counts.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Distribution of Price Ranges')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.show()
```

__Task-4__

```python
# --- TASK 4: ONLINE DELIVERY ---
print("\n--- Task 4: Online Delivery ---")
online_deliverycounts = df['Has Online delivery'].value_counts()
yes_percentage = (online_delivery_counts.get('Yes', 0) / total_restaurants) * 100
print(f"Percentage of restaurants with Online Delivery: {yes_percentage:.2f}%")

avg__rating_by_delivery = df.groupby('Has Online delivery')['Aggregate rating'].mean()
print("Average Rating comparison:")
print(avg__rating_by_delivery)
```
